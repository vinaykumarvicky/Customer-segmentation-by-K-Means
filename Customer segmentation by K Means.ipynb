```python
#Import the library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv(r"C:\Users\prana\Downloads\Customer_segmentation-_KMeans-main\Customer_segmentation-_KMeans-main\Mall_Customers.csv")
df.head()
```

```python
#Data Exploration
def basic_eda(df):
    print("----------HEAD--------")
    print(df.head())
    print("----------INFO-----------------")
    print(df.info())
    print("----------Describe-------------")
    print(df.describe())
    print("----------Columns--------------")
    print(df.columns)
    print("----------Data Types-----------")
    print(df.dtypes)
    print("-------Missing Values----------")
    print(df.isnull().sum())
    print("-------NULL values-------------")
    print(df.isna().sum())
    print("-----Shape Of Data-------------")
    print(df.shape)
```

```python
#Data Exploration
basic_eda(df)
```

```python
#Data Visualization
columns = ['Age', 'Annual Income (k$)','Spending Score (1-100)']
for i in columns:
    plt.figure()
    sns.distplot(df[i])
for i in columns:
    plt.figure()
    sns.kdeplot(df[i],shade=True,hue=df['Gender'])
for i in columns:
    plt.figure()
    sns.boxplot(data=df,x='Gender',y=df[i])
    
    
#We see that females vist the mall more than males.
```

```python
#Data Visualization
sns.pairplot(df.drop(columns='CustomerID'),hue='Gender')
```

```python
(df.groupby(['Gender'])['Age', 'Annual Income (k$)',
       'Spending Score (1-100)'].agg({np.mean,np.median})).T
```

```python
df.drop(columns='CustomerID').corr()
```

```python
#Segmentation using Annual Income
kss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df[['Annual Income (k$)']])
    kss.append(kmeans.inertia_)      # inertia is used for tell about the better quality of no.clusters
plt.plot(range(1,11),kss)
```

```python
#Segmentation using Annual Income k=3
clustering1 = KMeans(n_clusters=3)
clustering1.fit(df[['Annual Income (k$)']])
df['Income Cluster'] = clustering1.labels_
df.groupby('Income Cluster')['Age', 'Annual Income (k$)',
       'Spending Score (1-100)'].mean()
```

```python
#Segmentation using Annual Income and Spending Score
kss2=[]
for i in range(1,11):
    kmeans2=KMeans(n_clusters=i)
    kmeans2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
    kss2.append(kmeans2.inertia_)
plt.plot(range(1,11),kss2)
```

```python
#Segmentation using Annual Income and Spending Score k=5
clustering2 = KMeans(n_clusters=5)
clustering2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
df['Spending and Income Cluster'] =clustering2.labels_
df.groupby('Spending and Income Cluster')['Age', 'Annual Income (k$)',
       'Spending Score (1-100)'].mean()
```

```python
#Segmentation Using Annual Income and Spending Score and Visualization k=5
plt.figure(figsize=(15,8))
sns.scatterplot(data=df, x ='Annual Income (k$)',y='Spending Score (1-100)',
                hue='Spending and Income Cluster',style='Gender',palette='tab10')
```

We are most interested in 2 groups: purple and green.

Purple: low annual income - high spending score - mall need to support this group!

Green: high annual income - low spending score - need to get them to buy more!

```python
pd.crosstab(df['Spending and Income Cluster'],df['Gender'],normalize='index')
```
